```python
import pandas as pd
```

# Loading the data into a Pandas dataframe

```python
url = 'https://support.spatialkey.com/wp-content/uploads/2021/02/SacramentocrimeJanuary2006.csv'
read_csv_data = pd.read_csv(url)
```

```python
display(read_csv_data)
```

```python
# creating dataframe
dataframe = pd.DataFrame(read_csv_data, index=None)
 
# Count number of rows
rows_count = len(dataframe.axes[0])
 
# Count number of columns
columns_count = len(dataframe.axes[1])
```

# Count of Number of Rows and Columns

```python
# Count of Number of Rows and Columns
print("Number of rows: " + str(rows_count))
print("Number of columns: " + str(columns_count))
```

# Display the first 8 rows of the dataframe

```python
# Display the first 8 rows of the dataframe
print("The first 8 rows of the dataframe : ")
display(dataframe.head(8))
```

# Display the bottom 3 rows of the dataframe

```python
# Display the bottom 3 rows of the dataframe
print("The bottom 3 rows of the dataframe : ")
display(dataframe.tail(3))
```

# Displaying the transpose of the first 4 rows

```python
# Displaying the transpose of the first 4 rows
display(read_csv_data.head(4).T)
```
